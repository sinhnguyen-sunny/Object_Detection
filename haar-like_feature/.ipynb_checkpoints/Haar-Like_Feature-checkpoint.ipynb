## I.1 What is Haar-Like_Feature???

- Haar-Like Feature get the idea of “Haar wavelets” in recognition/dection problem [1]. 
- Specifically, it considers adjacent rectangular regions at a specific location in a detection window, sums up the pixel intensities in each region and calculates the difference between these sums. Features can consist of two, three or four rectangles as below:
    
    [2]

## I.2 Some Basic Ideas Using Haar-Like_Feature:

* *Detect Edge & Line:
    Ex: the region of the eyes is darker than the region of the cheeks ==> a common Haar feature for face detection is a set of two adjacent rectangles that lie above the eye and the cheek region. See the feature below:*
    
  
    [2]
* *HOG + Haar-Like Feature: encode an image ---> a vector*

**Especially, Applied into Viola–Jones object detection framework [3]:**
* **1.Haar Feature Selection --> 2.Creating an Integral Image --> 3.Adaboost Training --> 4.Cascading Classifiers**

## II.1 Prove the computing cost of the Haar-Like Feature algorithm

**Problem**
* **Assuming that:**
    * a detector has the resolution 24*24 pixels
    * subsequently placing rectangular features inside this detector in every position with every size possible 
    * These features include:


* **Question: How many is the exhaustive set of features????**

```python
#initial parameter
feature_dict = [(2,1), (1,2), (3,1), (1,3), (2,2)] #rectangular features
detector_size = 24 #the resolution sze of the detector
count = 0 #the initial number of exhaustive set of features
#solution
##Each shape of the rectangular features:
import sys 
for i in range(len(feature_dict)):
    size_x = feature_dict[i][0]
    size_y= feature_dict[i][1]
    ##Mutilples of basic shapes
    for width in range(size_x, detector_size - size_x +1,size_x):
        for height in range(size_y, detector_size - size_y+1, size_y):
            ##Each possiple position
            for x in range(0, int(detector_size - size_x)):
                for y in range(0, int(detector_size - size_y)):
                    count+=1
#                         print(count)
#                         sys.exit()
                                  
# Result
print(f'The exhaustive set of features: {count}')                                              
```
